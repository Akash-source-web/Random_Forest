```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
```

```python
# Import dataset
df=pd.read_csv('C:/Users/Akash Kundu/Desktop/ExcelR/Assingments/Fraud_check.csv')
```

### EDA

```python
df.info()
```

```python
# Add a column for Taxable.Income which contain data into binary details
df.loc[df['Taxable.Income']<=30000,'Taxable_Income']='Risky'
df.loc[df['Taxable.Income']>30000,'Taxable_Income']='Good'
```

```python
df
```

```python
df.drop('Taxable.Income',inplace=True,axis=1)
```

```python
df
```

```python
#convert catagorical data into numerical data after finding the unique values from the data
```

```python
df.Undergrad.unique()
df['Undergrad']=preprocessing.LabelEncoder().fit_transform(df['Undergrad'])
```

```python
df['Marital.Status'].value_counts()
```

```python
df['Marital.Status']=preprocessing.LabelEncoder().fit_transform(df['Marital.Status'])
```

```python
df.Urban.unique()
```

```python
df['Urban']=preprocessing.LabelEncoder().fit_transform(df['Urban'])
```

```python
df.Taxable_Income.unique()
```

```python
df
```

```python
# Divide the data into target and predictor
columns=list(df.columns)
predictor=columns[0:5]
target=columns[-1]
```

```python
predictor
```

```python
target
```

```python
X=df[predictor]
Y=df[target]
```

### Model Building

```python
rf=RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion='entropy')
```

```python
rf
```

### Model Prediction

```python
rf.fit(X,Y)
```

```python
rf.estimators_
```

```python
rf.classes_
```

```python
df['rf_predict']=rf.predict(X)
```

```python
#Creating Confusion matrix for the prediction
confusion_matrix(df['Taxable_Income'],df['rf_predict'])
```
